```python
years = range(1880,2011)
pieces = []
columns = ['name','sex','births']

for year in years:
    frame = pd.read_csv(f'/home/jovyan/pydata-book/ch02/names/yob{year}.txt',names = columns)
    frame['year'] = year
    pieces.append(frame)
    
names = pd.concat(pieces, ignore_index = True)
```

```python
names.head()
```

```python
total_births = pd.pivot_table(names,values = 'births',index = 'year',columns = 'sex', aggfunc = 'sum')
total_births.head()
```

```python
def add_prop(group):
    births = group.births.astype(float)
    group['prop'] = births / births.sum()
    return group
```

```python
names = names.groupby(['year','sex']).apply(add_prop)
```

```python
np.allclose(names.groupby(['year','sex']).prop.sum(),1)
```

```python
(names.groupby(['year','sex'])).head()
```

```python
#Years 1880 and 1881 have only about 1900 names, so top 1000 M and F will not give a full 2000 list
def get_top1000(groups):
    return groups.sort_index(axis = 0, level='births', ascending = True)[:1000]
grouped = names.groupby(['year','sex'])
top_1000 = grouped.apply(get_top1000)
```

```python
pieces = []
for year, groups in names.groupby(['year','sex']):
    x = groups.sort_index(axis = 0, level = 'births', ascending = True)[:1000]
    pieces.append(x)
top_1000 = pd.concat(pieces, ignore_index = False)
```

```python
boys = top_1000[top_1000['sex'] == 'M']
girls = top_1000[top_1000['sex'] == 'F']
```

```python
total_births = pd.pivot_table(top_1000, values = 'births', index = 'year', columns = 'name', aggfunc = 'sum')
```

```python
def get_quantile(group, q = 0.5):
    group = group.sort_index(axis = 0, level = 'prop', ascending = True)
    return group.prop.cumsum().searchsorted(q) + 1

diversity = top_1000.groupby(['year','sex']).apply(get_quantile)
diversity = diversity.unstack('sex')
```
