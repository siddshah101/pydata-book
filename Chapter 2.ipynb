```python
def get_counts(sequence):
    counts = {}
    for x in sequence:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts
```

```python
get_counts(time_zones)
```

```python
from collections import defaultdict
```

```python
def get_counts2(sequence):
    counts = defaultdict(int)
    for x in sequence:
        counts[x] += 1
    return counts
```

```python
get_counts2(time_zones)
```

```python
def top_counts(sequence,n=10):
    value_key_pairs = [(counts,tz) for tz, count in sequence.items()] #Simply reframing to put counts in front of the respective time zone
    value_key_pairs.sort()
    return value_key_pairs[-n:]
```
